```python
# Cloumns : [Age, Math Marks, Science Marks]
data = np.array([
    [18,85,78],  #Student 1
    [19,92,88],  #Student 2
    [17,76,95],  #Student 3
    [18,65,70],  #Student 4
    [20,90,85],  #Student 5
])
```

```python
#Get the shape of the matrix.
# np.shape(data)
data.shape
```

```python
#Find the average age of students.
# a = data[:,0]
# a
# np.mean(a)
np.mean(data[:,0])
```

```python
# Extract Math marks of all students.
math_marks = data[:,1]
math_marks
```

```python
#find the highest Science marks.
np.max(data[:,2])
```

```python
#Get details of the student who scored more than 90 in Math.
data[data[:,1] > 90]
```

```python
#Increase Math marks of all students by 5.
data[:,1] +=5
data
```

```python
#Find how many students are younger then 19.
# data[data[:,0]<19]
len(data[data[:,0]<19])
```

```python
#Calculate the average marks in each subject (column-wise mean).
# data[:,1:]
np.mean(data[:,1:],axis=0)
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```

```python
# Replace all Science marks < 75 with 0.
data[:,2][data[:,2]<75] = 0
```
