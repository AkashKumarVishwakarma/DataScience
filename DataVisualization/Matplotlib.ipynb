**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(x**2,x**2)

plt.subplot(2,2,3)
plt.plot(y,x)
```

**Object Oriented way**

```python
fig = plt.figure()
# fig.add_axes([0.1,0.1,0.2,0.8])
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.1,1,1])
axis2.plot(x,y)
```

```python
fig = plt.figure()
# fig.add_axes([0.1,0.1,0.2,0.8])
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,.3,.3])
axis2.plot(x,y)
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi
keywords arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch(pixel per inch).

```python
fig = plt.figure(figsize=(12,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,.3,.3])
axis2.plot(x,y)
```

```python
xaxis = np.linspace(0,5,11)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(xaxis,xaxis+1,color="red",linewidth = 0.25)
ax.plot(xaxis,xaxis+2,color="red",linewidth = 0.50)
ax.plot(xaxis,xaxis+3,color="red",linewidth = 1.00)
ax.plot(xaxis,xaxis+4,color="red",linewidth = 2.00)

#Possible linestype options '-','--','-.' ':' 'steps'
ax.plot(xaxis,xaxis+5,color="green",lw=3, linestyle= '-')
ax.plot(xaxis,xaxis+6,color="green",lw=3, linestyle= '-.')
ax.plot(xaxis,xaxis+7,color="green",lw=3, linestyle= ':')

#custom dash
line, = ax.plot(xaxis,xaxis+8,color='black',lw=1.50)
line.set_dashes((5,50,15,50)) # format: line length, space length, ...

#possible marker symbols marker = '+','*'
ax.plot(xaxis, xaxis+9, color='blue',lw=3, ls='-',marker='+')
ax.plot(xaxis, xaxis+10, color='blue',lw=3, ls='--',marker='*')
```
