```python
import plotly.data as pldata

tipsN = pldata.tips()

fig = px.line(tips, y='total_bill')
fig.show()
```

```python
# tips.groupby('day')['tip'].mean().iplot(kind = 'bar')

# Plotly Express
avg_tip = tips.groupby('day', as_index=False)['tip'].mean()

fig = px.bar(
    avg_tip,
    x='day',
    y='tip',
    title = 'Average Tip by day'
)
fig.show()
```
