```python
import numpy as np
import pandas as pd
```

**GroupBy**

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,180,200,300],
    'Quantity':[10,15,12,18,8,20,15,25],
    'Date':pd.date_range('2023-01-01',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Group By Category and calculate the sum of Sales
cat = df.groupby('Category')
cat
```

```python
for i,v in cat:
    print(i)
    print(v)
```

```python
cat_sum = df.groupby('Category')['Sales'].sum()
cat_sum
```

```python
# Group by Store calculate the sum of Sales
sales_sum = df.groupby('Store')['Sales'].sum()
sales_sum
```

```python
#Group by multiple columns
# Group by Category and Store

cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

**Aggregation**

```python
df['Sales'].mean()
```

```python
#median mode min max count std 
```

```python
df['Sales'].agg(['sum','mean','min','max','count','std','median'])
```

```python
df['Sales'].mod(7)
```

```python
#Count Records per Category
df.groupby('Category').size()
```

```python
#Highest Sale in Each Category
df.groupby('Category')['Sales'].max()
```

```python
#Sort Groups by Total Sales
df.groupby('Category')['Sales'].sum().sort_values(
    ascending = False
)
```

```python
#Top Selling Store
df.groupby('Store')['Sales'].sum()
```

```python
#Range(custom Aggregation)
df.groupby('Category')['Sales'].agg(
    lambda x: x.max() - x.min()
)
```

```python
#Filter Groups (show only Sales > 600)

df.groupby('Category').filter(
    lambda x: x['Sales'].sum() > 600
)
```

```python
# Category ke ander sabse zyada sale wali row
idx = df.groupby('Category')['Sales'].idxmax()
df.loc[idx]
```
