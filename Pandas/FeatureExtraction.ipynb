```python
# Make a new Column for episode count
# df.loc[2]
df.loc[2]['Title']
```

```python
txt = df.loc[2]['Title']

def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ')':
            check = False
            return data
        if check == True:
            data += i
        if i == '(':
            check = True
        
            
```

```python
df['Title'].apply(extract_episodes)
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps","")
```
